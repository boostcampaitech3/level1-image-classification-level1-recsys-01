```python
import pandas as pd
import os
import shutil
```

```python
data_list = pd.read_csv(r"/opt/ml/input/data/train/train_edit.csv")
data_list.head()
```

```python
# 라벨링 폴더 생성 (label)
os.mkdir(r"/opt/ml/input/data/train/label")
```

```python
# 라벨링 폴더 생성 (0 ~ 17)
for i in range(18):
    os.mkdir(r"/opt/ml/input/data/train/label/{}".format(str(i)))
```

```python
path = "/opt/ml/input/data/train/images"
```

```python
file_lst = os.listdir(path)
files = []
for i in file_lst:
    if '._' not in i:
        files.append(i)
files[:10]
```

```python
# 라벨링 함수 생성 
# shutil.copy() : 파일 복사하는 함수


def labeling(srcPath, srcFile, desPath, desFile, gd, age):
    if srcFile[0] == 'm':
        if age < 30:
            if gd == "male":
                shutil.copy(srcPath + srcFile, desPath + "0/" + desFile)
            else:
                shutil.copy(srcPath + srcFile, desPath + "3/" + desFile)
        elif age < 60:
            if gd == "male":
                shutil.copy(srcPath + srcFile, desPath + "1/" + desFile)
            else:
                shutil.copy(srcPath + srcFile, desPath + "4/" + desFile)
        else:
            if gd == "male":
                shutil.copy(srcPath + srcFile, desPath + "2/" + desFile)
            else:
                shutil.copy(srcPath + srcFile, desPath + "5/" + desFile)
    elif srcFile[0] == 'i':
        if age < 30:
            if gd == "male":
                shutil.copy(srcPath + srcFile, desPath + "6/" + desFile)
            else:
                shutil.copy(srcPath + srcFile, desPath + "9/" + desFile)
        elif age < 60:
            if gd == "male":
                shutil.copy(srcPath + srcFile, desPath + "7/" + desFile)
            else:
                shutil.copy(srcPath + srcFile, desPath + "10/" + desFile)
        else:
            if gd == "male":
                shutil.copy(srcPath + srcFile, desPath + "8/" + desFile)
            else:
                shutil.copy(srcPath + srcFile, desPath + "11/" + desFile)

    elif srcFile[0] == "n":
        if age < 30:
            if gd == "male":
                shutil.copy(srcPath + srcFile, desPath + "12/" + desFile)
            else:
                shutil.copy(srcPath + srcFile, desPath + "15/" + desFile)
        elif age < 60:
            if gd == "male":
                shutil.copy(srcPath + srcFile, desPath + "13/" + desFile)
            else:
                shutil.copy(srcPath + srcFile, desPath + "16/" + desFile)
        else:
            if gd == "male":
                shutil.copy(srcPath + srcFile, desPath + "14/" + desFile)
            else:
                shutil.copy(srcPath + srcFile, desPath + "17/" + desFile)
```

```python
# 폴더별 이미지별 라벨링 진행
# 새로 저장된 파일명 : 폴더명_파일명

file_cnt = 0
folder_cnt = 0
for i in range(len(data_list)):
    gd, age = data_list.iloc[i]['gender'], data_list.iloc[i]['age']
    files_ = os.listdir(path + '/' + data_list.iloc[i]['path'])
    srcpath = "/opt/ml/input/data/train/images/" + data_list.iloc[i]['path'] + '/'
    despath = "/opt/ml/input/data/train/label/"
    desfile = data_list.iloc[i]['path']
    for i in files_:
        labeling(srcpath, i, despath, desfile + '_' + i, gd, age)
        file_cnt += 1
    
    folder_cnt += 1

    print(file_cnt, folder_cnt)
```
